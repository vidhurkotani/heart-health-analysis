# classification of data 

```python
x=heart.iloc[:,:-1]
y=heart.iloc[:,-1]
```

```python
x
```

```python
y
```

# classification-- normalising the labels

```python
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

# dividing into train and test data


```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# knn model

```python
#fitting classifier to training set
```

```python
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
#prediction on test data
```

```python
y_pred=clf.predict(x_test)
print("predicted heart disease")
print(y_pred)
```

```python
print("actuall heart disease:")
print(y_test.values)
```

```python
#accuracy score on train and test
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\n Accuracy score:%f"%(accuracy_score(y_test,y_pred)*100))
print("recall_score:%f"%(recall_score(y_test,y_pred)*100))
print("ROC score:%f\n"%(roc_auc_score(y_test,y_pred)*100))
print(confusion_matrix(y_test,y_pred))
```
